```python
#Functions break code into smaller chunks and are great for defining actions that a program will execute several 
# times throughout your code.
#The process for executing a function is in three steps: 1. The function is called (arguments are passed as input)
#2. The function executes (some action is performed with the arguments).3. The function returns,(original function call is replaced with the return value.)
```

```python
def multiply (x,y):
    product = x * y
    return product
    
num = multiply (2, 4)
print (num)
```

```python
#convert cels to far (temperature)
def conv_cel_to_far (celsius):
    #Convert Celsius to Fahrenheit.
    Formula: Fahrenheit = Celsius * 9/5 + 32
    return celsius * 9 / 5 + 32

def conv_far_to_cel (fahrenheit):
    #Convert Fahrenheit to Celsius.
    Formula: celsuis = (fahrenheit - 32) * 5/9
    return fahrenheit - 32 * 5 / 9
```

```python
def main():
    # Prompt user for Fahrenheit input
    fahrenheit = float(input("Enter temperature in degrees Fahrenheit: "))
    celsius_converted = convert_far_to_cel(fahrenheit)
    print(f"{fahrenheit:.2f}°F is equal to {celsius_converted:.2f}°C")
```
