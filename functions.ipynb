```python
#convert cels to far (temperature)
def conv_celsius_to_farenheit (c):
    #Convert Celsius to Fahrenheit.
    return (c * 9 / 5) + 32

def conv_farenheit_to_celsuis (f):
    #Convert Fahrenheit to Celsius.
    return (f - 32) * 5 / 9
```

```python
def main():
    # Prompt user for Fahrenheit and celsius input
    print("Temperature converter: celsuis ↔ fahrenheit")
    print("1. celsius to fahrenheit")
    print("2. fahrenheit to celsius")

choice = input("enter your choice (1 or 2): ").strip()

    
```

```python
if choice == '1':
    try:
        c = float(input("enter temperature in celsius: "))
        f = conv_celsius_to_farenheit (c)
        print(f"{c}°C = {f:.2f}°F")
    except ValueError:
        print("Please enter a valid number.")
        
elif choice == '2':
    try:
        f = float(input("enter temperature in fahrenheit: "))
        c = conv_farenheit_to_celsuis (f)
        print(f"{f}°F = {c:.2f}°C")
    except ValueError:
        print("Please enter a valid number.")
```
